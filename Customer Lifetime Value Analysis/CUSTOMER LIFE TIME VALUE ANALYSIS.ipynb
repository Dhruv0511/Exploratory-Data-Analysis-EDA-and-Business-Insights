# CUSTOMER LIFE TIME VALUE ANALYSIS

```python
import pandas as pd
import plotly.graph_objs as go
import plotly.express as px
import plotly.io as pio
pio.templates.default = 'plotly_white'
```

```python
clv_data = pd.read_csv('customer_acquisition_data.csv')
```

```python
clv_data.head()
```

Letâ€™s start by visualizing the distribution of acquisition cost and revenue generated by the customer using histograms

```python
fig = px.histogram(clv_data, x='cost', nbins = 20, title = 'Distribution of Acquisition cost')

fig.show()
```

```python
fig = px.histogram(clv_data, x= 'revenue', nbins= 20, title = 'Distribution by revenue')

fig.show()
```

Now letâ€™s compare the cost of acquisition across different channels and identify the most and least profitable channels

```python
cost_by_channel = clv_data.groupby('channel')['cost'].mean().reset_index()

fig = px.bar(cost_by_channel, x='channel', y='cost', 
             title = 'Customer Acquisition Cost By Channel')

fig.show()
```

```python
conversion_by_channel = clv_data.groupby('channel')['conversion_rate'].mean().reset_index()

fig = px.bar(conversion_by_channel, x = 'channel', y='conversion_rate',
            title='Conversion Rate by Channel')

fig.show()
```

```python
revenue_by_channel = clv_data.groupby('channel')['revenue'].sum().reset_index()

fig = px.pie(revenue_by_channel, values = 'revenue', names = 'channel',
            title = 'Total Revenue by Channel', hole = 0.6,
            color_discrete_sequence=px.colors.qualitative.Pastel)

fig.show()
```

```python
clv_data['roi'] = clv_data['revenue']/clv_data['cost']

roi_by_channel = clv_data.groupby('channel')['roi'].mean().reset_index()

fig = px.bar(roi_by_channel, x= 'channel', y='roi',
            title = 'Return on Investment by Channel')

fig.show()
```

```python
clv_data['cltv'] = (clv_data['revenue']-clv_data['cost']) * clv_data['conversion_rate'] / clv_data['cost']

channel_cltv = clv_data.groupby('channel')['cltv'].mean().reset_index()

fig = px.bar(channel_cltv, x='channel', y='cltv',
            title = 'Customer Life Time Value By Channel', color='channel')

fig.update_xaxes(title = 'Channel')
fig.update_yaxes(title = 'CLTV')

fig.show()
```
